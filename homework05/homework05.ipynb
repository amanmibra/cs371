## Part 1: Numerical Differentiation

```python
import numpy as np


def function(x):
    return np.sin(x)


def derivative(x):
    return np.cos(x)


def forward_difference(f, x, h=1.e-5):
    assert h > 0, 'h must be positive'
    return (f(x + h) - f(x)) / h


def error(method, f, d, x):
    h_values = np.logspace(-16, -1, num=101)
    return [np.abs(method(f, x, h) - d(x)) for h in h_values], h_values, method.__name__
```

### Problem 1.1

```python
import matplotlib.pyplot as plt

y0, x0, _ = error(forward_difference, function, derivative, 0)
y1, x1, _ = error(forward_difference, function, derivative, 1)

plt.loglog(x0, y0, label="x = 0")
plt.loglog(x1, y1, label="x = 1")
plt.legend()
plt.xlabel('h values')
plt.ylabel('derivative aprox error')
plt.show()
```

```python
def central_difference(f, x, h=1.e-5):
    assert h > 0, 'h must be positive'
    return ((f(x + h) - f(x - h)) / (2 * h))

def four_point_difference(f, x, h=1.e-5):
    assert h > 0, 'h must be positive'
    return (-f(x + 2*h) + 8*f(x+h) - 8*f(x-h) + f(x-2*h)) / (12*h)

y0, x0, label0 = error(forward_difference, function, derivative, 1)
y1, x1, label1 = error(central_difference, function, derivative, 1)
y2, x2, label2 = error(four_point_difference, function, derivative, 1)

plt.loglog(x0, y0, label=label0)
plt.loglog(x1, y1, label=label1)
plt.loglog(x2, y2, label=label2)
plt.legend()
plt.xlabel('h values')
plt.ylabel('derivative aprox error')
plt.show()
```

## Part 3: The Logistic-Regression Loss

### Problem 3.1

```python
#ylogp(1y)log(1p)
def loss(y, p):
    return -y * np.log(p) - (1 - y)*np.log(1-p)

def score(a):
    return 1/(1 + np.exp(-a))

def regression_loss(y, alpha):
    return [loss(y, score(a)) for a in alpha]

x = np.arange(-3,3, 0.1)
y0 = regression_loss(0, x)
y1 = regression_loss(1, x)

plt.plot(x,y0, label="y = 0")
plt.plot(x,y1, label="y = 1")
plt.legend()
plt.show()
```
